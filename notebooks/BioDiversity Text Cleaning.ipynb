```python
import os
import pandas as pd

data_path = "text"

# Create the Taxonomy directory if it doesn't exist
taxonomy_dir = os.path.join(data_path, "taxonomy")
if not os.path.exists(taxonomy_dir):
    os.mkdir(taxonomy_dir)

# Load the DataFrame (Replace this with your method of loading the DataFrame)
# df = pd.read_csv("your_data.csv")

# Select the first 100 species from the DataFrame
selected_species = df.head(100)

for index, row in selected_species.iterrows():
    species_name = row["SpeciesName"]
    species_dir = os.path.join(data_path, species_name)

    # Create the species directory if it doesn't exist
    if not os.path.exists(species_dir):
        os.mkdir(species_dir)

    # Create the Taxonomy directory within the species directory
    species_taxonomy_dir = os.path.join(species_dir, "taxonomy")
    if not os.path.exists(species_taxonomy_dir):
        os.mkdir(species_taxonomy_dir)

    # Create a text file to store taxonomy information
    taxonomy_file_path = os.path.join(species_taxonomy_dir, "taxonomy.txt")
    with open(taxonomy_file_path, "w") as taxonomy_file:
        taxonomy_file.write("Root: " + str(row["Root"]) + "\n")
        taxonomy_file.write("Kingdom: " + str(row["Kingdom"]) + "\n")
        taxonomy_file.write("Phylum: " + str(row["Phylum"]) + "\n")
        taxonomy_file.write("Class: " + str(row["Class"]) + "\n")
        taxonomy_file.write("Order: " + str(row["Order"]) + "\n")
        taxonomy_file.write("Family: " + str(row["Family"]) + "\n")
        taxonomy_file.write("Genus: " + str(row["Genus"]) + "\n")
        taxonomy_file.write("Species: " + str(row["Species"]) + "\n")
```

```python
import os
import pandas as pd

data_path = "text"

# Create the Data directory if it doesn't exist
data_dir = os.path.join(data_path, "data")
if not os.path.exists(data_dir):
    os.mkdir(data_dir)

# Load the DataFrame (Replace this with your method of loading the DataFrame)
# df = pd.read_csv("your_data.csv")

# Select the first 100 species from the DataFrame
selected_species = df.head(100)

# List of attributes to store in the text files
attributes = ["Overview", "Size", "Morphology", "Behavior", "Distribution", "Habitat", "HabitatDetails"]

for index, row in selected_species.iterrows():
    species_name = row["SpeciesName"]
    species_dir = os.path.join(data_path, species_name)

    # Create the species directory if it doesn't exist
    if not os.path.exists(species_dir):
        os.mkdir(species_dir)

    # Create the Taxonomy directory within the species directory
    species_text_data = os.path.join(species_dir)
    if not os.path.exists(species_text_data):
        os.mkdir(species_text_data)

    # Create and write text files for each attribute
    for attr in attributes:
        attr_file_path = os.path.join(species_text_data, f"{attr}.txt")
        with open(attr_file_path, "w") as attr_file:
            attr_file.write(str(row[attr]))
```

```python
category_counts = {}

for entry in df['Conservation Status']:
    categories = entry.split("\n")
    for category in categories:
        category = category.strip()  # Remove leading/trailing whitespaces
        if category:  # Skip empty strings
            if category not in category_counts:
                category_counts[category] = 1
            else:
                category_counts[category] += 1

for category, count in category_counts.items():
    print(f"{category}: {count} occurrences")
```

```python
red_list_categories = {
    "Not Evaluated", "Data Deficient", "Least Concern",
    "Near Threatened", "Vulnerable", "Endangered",
    "Critically Endangered", "Extinct in the Wild", "Extinct"
}

category_counts = {category: 0 for category in red_list_categories}

for entry in df['Conservation Status']:
    categories = entry.split("\n")
    for category in categories:
        category = category.strip()
        for red_list_category in red_list_categories:
            if red_list_category in category:
                category_counts[red_list_category] += 1

for category, count in category_counts.items():
    print(f"{category}: {count} occurrences")
```

```python
import os
import pandas as pd

data_path = "text"

# Load the DataFrame (Replace this with your method of loading the DataFrame)
# df = pd.read_csv("your_data.csv")

# Select the first 100 species from the DataFrame
selected_species = df.head(100)

# Red list categories
red_list_categories = {
    "Not Evaluated", "Data Deficient", "Least Concern",
    "Near Threatened", "Vulnerable", "Endangered",
    "Critically Endangered", "Extinct in the Wild", "Extinct"
}

# Initialize category counts
category_counts = {category: 0 for category in red_list_categories}

for index, row in selected_species.iterrows():
    species_name = row["SpeciesName"]
    species_dir = os.path.join(data_path, species_name)
    species_text_dir = os.path.join(species_dir)

    # Create the species text directory if it doesn't exist
    if not os.path.exists(species_text_dir):
        os.makedirs(species_text_dir)

    # Extract and store the red list status
    conservation_status = row["Conservation Status"]
    red_list_status = ""
    for category in red_list_categories:
        if category in conservation_status:
            red_list_status = category
            break

    # Store the red list status in a text file
    red_list_status_file_path = os.path.join(species_text_dir, "RedListStatus.txt")
    with open(red_list_status_file_path, "w") as red_list_status_file:
        red_list_status_file.write(red_list_status)

    print(f"Stored Red List Status for {species_name}: {red_list_status}")
```
