```python
bs = 8
train_loader = DataLoader(train_data, batch_size=bs, shuffle=True)
valid_loader = DataLoader(val_data, batch_size=bs, shuffle=False)
```

```python
# Define model architecture
class DistilRobertaForClassification(nn.Module):
    def __init__(self, num_labels):
        super(DistilRobertaForClassification, self).__init__()
        self.distilroberta = AutoModelForSequenceClassification.from_pretrained('distilroberta-base', num_labels=num_labels)

    def forward(self, input_ids, attention_mask):
        logits = self.distilroberta(input_ids=input_ids, attention_mask=attention_mask)[0]
        return logits

model = DistilRobertaForClassification(num_labels=83)
model.to(device);
```

```python
num_epochs = 10
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
criterion = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
```

```python
def training(batch, model, device):
  input_ids = batch['input_ids'].to(device)
  attention_mask = batch['attention_mask'].to(device)
  labels = batch['labels'].to(device)
  logits = model(input_ids, attention_mask)
  loss = criterion(logits, labels)
  return logits, loss, labels
```

```python
def validation(batch, model, device):
  input_ids = batch['input_ids'].to(device)
  attention_mask = batch['attention_mask'].to(device)
  labels = batch['labels'].to(device)
  logits = model(input_ids, attention_mask)
  loss = criterion(logits, labels)
  return logits, loss, labels
```

```python
for epoch in range(num_epochs):
    # Train loop
    model.train()
    train_loss = 0
    train_acc = 0
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training"):
        optimizer.zero_grad()
        logits, loss, labels = training(batch, model, device)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        train_acc += torch.sum(preds == labels).item() / labels.shape[0]
    train_loss /= len(train_loader)
    train_acc /= len(train_loader)

    # Validation loop
    model.eval()
    val_loss = 0
    val_acc = 0
    for batch in tqdm(valid_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Validation", leave=False):
        logits, loss, labels = validation(batch, model, device)
        val_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        val_acc += torch.sum(preds == labels).item() / labels.shape[0]
    val_loss /= len(valid_loader)
    val_acc /= len(valid_loader)

    scheduler.step()

    # Print the epoch statistics
    tqdm.write(f'Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f} - Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}')

    # Save the model
    torch.save(model.state_dict(), f"models/epoch-{epoch+1}:acc-{val_acc:.4f}.pth")
```
