```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "/content/drive/MyDrive/Colab Projects/BioDiversity Project/Saved_LLM_Model_V3"  # Replace with your model name
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Import necessary libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Define a conversation history
conversation_history = []

# Start a conversation loop
while True:
    # User input
    user_input = input("You: ")

    # Add user input to the conversation history
    conversation_history.append(user_input)

    # Encode the conversation history
    input_ids = tokenizer.encode(" ".join(conversation_history), return_tensors="pt")

    # Generate a response from the model
    with torch.no_grad():
        output = model.generate(input_ids, max_length=500, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)

    # Decode and print the model's response
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    print(f"Model: {response}")

    # Add model's response to the conversation history
    conversation_history.append(response)
```

```python
# Import necessary libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# User input
user_input = input("You: ")

# Encode the user input
input_ids = tokenizer.encode(user_input, return_tensors="pt")

# Generate a response from the model
with torch.no_grad():
    output = model.generate(input_ids, max_length=100, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)

# Decode and print the model's response
response = tokenizer.decode(output[0], skip_special_tokens=False)
print(f"Model: {response}")
```
