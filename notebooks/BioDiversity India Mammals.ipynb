```python
df = pd.read_csv(r"/content/drive/MyDrive/BioDiversity Project/data/Mammals - Sheet1.csv")
```

```python
# Function to extract the first two words from a string
def extract_species_name(species):
    if isinstance(species, str):
        words = species.split()
        if len(words) >= 2:
            return ' '.join(words[:2])
    return ""

# Extract the first two words from the Species column and fill the SpeciesName column
df['SpeciesName'] = df['Species'].apply(extract_species_name)

# Print the updated DataFrame
print(df)
```
